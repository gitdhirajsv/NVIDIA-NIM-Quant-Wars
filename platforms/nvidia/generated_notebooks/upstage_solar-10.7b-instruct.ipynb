# Results for model: upstage/solar-10.7b-instruct

```
import polars as pl
from xgboost import XGBRegressor

# step 1: Load 'train.parquet'
df = pl.read_parquet('./jane_street_data/train.parquet')

# step 2: Feature engineering
# Calculate top 15% quantile of 'feature_00'
top_quartile = df.select(["date_id", "feature_00", "responder_6"]).groupby("date_id").quantile(0.85, "feature_00").alias("top_quartile")

# Join quantile calculation with original dataframe
df = df.join(
    top_quartile,
    on=["date_id"],
    how="left",
)

# step 3: Train XGBoost Regressor
xgb = XGBRegressor()
xgb.fit(df[df.columns not in {"responder_6"}], df["responder_6"])
```
