# Results for model: meta/llama3-8b-instruct

```
import polars as pl
import xgboost as xgb
import numpy as np

train_df = pl.read_parquet('./jane_street_data/train.parquet')

train_df = train_df.with_column(pl.col('feature_00').rolling(window=30).apply(lambda x: x.quantile(0.85)))

train_df = train_df.with_column(pl.col('responder_6').astype(pl.Float32))

xgb_model = xgb.XGBRegressor()
xgb_model.fit(train_df.select(['feature_00', 'responder_6']).to_pandas(), train_df.select(['responder_6']).to_pandas())
```
