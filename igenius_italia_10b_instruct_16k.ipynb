# Results for model: igenius/italia_10b_instruct_16k

```
import polars as pl
from xgboost import XGBRegressor

# Load dataset
df = pl.read_table('./jane_street_data/train.parquet')

# Feature Engineering
df['feature_00_quantiles'] = df.apply_expr('feature_00 > percentile(feature_00, 0.85)').astype(int)

# Train an XGBoost Regressor
model = XGBRegressor()
model.fit(df[['feature_00_quantiles']], df['responder_6'])

# Save model
model.save_model('xgboost_model.mdl')
```
